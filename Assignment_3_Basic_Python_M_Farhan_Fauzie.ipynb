```python
# Import library
import pandas as pd
import numpy as np

# Import Dataset
dataset = pd.read_csv('https://raw.githubusercontent.com/hadimaster65555/dataset_for_teaching/main/dataset/linear_regression_simulation_data/linear-regression.csv')
dataset

headers =  ["X", "Y"]
dataset.columns = headers
dataset

# Import Visualization
import matplotlib.pyplot as plt

plt.plot(dataset.X, dataset.Y, 'o', label = 'random data')
plt.legend()
plt.show()


```

2. Buatlah program untuk menghitung regresi linier menggunakan **metode invers matriks**

  x = (A^T * A)^(-1) * A^T * y
dimana :
  - x adalah koefisien dari model regresi yang ingin dicari
  - A adalah matriks data
  - A^T adalah transpose dari A
  - A^(-1) adalah invers dari matriks A
  - y adalah variabel target

```python
# Nilai Kolom
x = dataset.X.values
y = dataset.Y.values

# Invers Matriks
x_matrix = np.vstack((np.ones(len(x)), x)).T
x_invers = np.linalg.inv(x_matrix.T.dot(x_matrix)).dot(x_matrix.T).dot(y)

y_pred = x_matrix.dot(x_invers)

# Sebaran
plt.plot(x, y_pred, color='red',label="fitted line");
plt.scatter(x, y,label="data");

# Text
plt.text(0.35,2, r"$\beta_0 = 3.992$", fontsize=14, color="b")
plt.text(0.35,1.5, r"$\beta_1 = 2.981$", fontsize=14, color="b")
plt.legend()
plt.show()


```

3. Buatlah program untuk menghitung regresi linier menggunakan method **linalg.lstsq dari numpys**

Untuk poin 2 dan 3 mentee wajib menvisualkan hasil dari model seperti contoh berikut:

image.png



```python
# Nilai Kolom
x = dataset.X
y = dataset.Y

# Rumus Linear
A = np.array([x, np.ones(x.shape[0])])
w = np.linalg.lstsq(A.T, y, rcond=None)[0]

# Plot Sebaran dan Garis Regresi
line = w[0]*x + w[1]  
plt.plot(x, line, 'r-',label="fitted line")
plt.plot(x, y, 'o',label='data')

# Text
plt.text(0.35,2, r"$\beta_0 = 3.992$", fontsize=14, color="b")
plt.text(0.35,1.5, r"$\beta_1 = 2.981$", fontsize=14, color="b")
plt.legend()
plt.show()
```
